## Random Forest 

I decided to go with the Random Forest model for now, as it's a well balanced model that can produce good results.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Initialize the Random Forest model
random_forest_model = RandomForestClassifier(n_estimators=50, random_state=42)

# Train the model on the training data
random_forest_model.fit(X_train, y_train)
```

```python
# Make predictions on the testing data
y_pred = random_forest_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Get a classification report
print(classification_report(y_test, y_pred))
```

The model performed poorly, but as we can see from the support column, the amount of columns I picked with actual fraud data was very little, so we're going to edit the data to choose more columns that were in fact fraudulent.

```python
df['is_fraud'].value_counts()
```

There's only around 7500 fraudelent columns total so we're going to ensure that all of those get picked, as well as about 50,000 non-fraud columns, since 100,000 looked to be overkill here. 

```python
# Selecting all columns where 'is_fraud' equals 1
fraudulent_data = df[df['is_fraud'] == 1]

# Selecting 25,000 random columns where 'is_fraud' equals 0
non_fraudulent_data = df[df['is_fraud'] == 0].sample(n=25000, random_state=42)
```

```python
combined_data = pd.concat([fraudulent_data, non_fraudulent_data])
combined_data
```

Now let's try again.

```python
df_RF = combined_data
```

```python
df_RF['is_fraud'].value_counts()
```

```python
# Split the data
X = df_RF.drop('is_fraud', axis=1)
y = df_RF['is_fraud']
```

```python
# Train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Initialize the Random Forest model
random_forest_model = RandomForestClassifier(n_estimators=50, random_state=42)

# Train the model on the training data
random_forest_model.fit(X_train, y_train)
```

```python
# Make predictions on the testing data
y_pred = random_forest_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Get a classification report
print(classification_report(y_test, y_pred))
```

Even though the accuracy of the model decreased slightly, it did a MUCH better job on the columns that were actually fraudulent. 

The model's weakest point was it's recall on fraudelent purchases. This is a concern because that is the most essential thing to know when it comes to fraudulent activity. A missed positive means lost revenue. The high accuracy on non-fraud purchases is important too, preventing a reduction in customer satisfaction with false accusations.

## Hyperparameter Tuning

The Random Forest Classifer performed well so we will now tune the model to try and increase the overall accuracy. 

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Define hyperparameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

(Initally included "none" in max_depth but that would make this take too long.)

```python
# Create Grid Search object
grid_search = GridSearchCV(estimator=random_forest_model, param_grid=param_grid, cv=5, scoring='accuracy')

# Fit the grid search to the data
grid_search.fit(X_train, y_train)
```

```python
# Get the best model
best_params = grid_search.best_params_
best_params
```

```python
best_model = grid_search.best_estimator_
```

```python
best_model
```

```python
# Make predictions on the test set
y_pred = best_model.predict(X_test)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Get a classification report
print(classification_report(y_test, y_pred))
```

## Model Evaluation

As we can see from the results, tuning the parameters helped the model slightly, but not by much. Higher depth and estimators did make a positive impact though. 96.5% accuracy for a model is very good overall.

Max depth is probably the most impactful of the parameters and would probably be the best thing to tweak, then estimators.

## Creating the Dash app part 1

Here I tried to re-create the most releveant looking dash application from the Dash documentation:

```python
# Process to create the dash app, including pre-processing, the layout, callback, and output.

def preprocess_features(features):
    # Convert features to appropriate data types
    amount = float(features[0])
    zip_code = str(features[1])
    city_population = int(features[2])
    processed_features = [amount, zip_code, city_population]

    return processed_features

def predict(amount, zip_code, city_population):
    features = [amount, zip_code, city_population]
    processed_features = preprocess_features(features)
    prediction = model.predict([processed_features])[0]  # Extract prediction from array
    return prediction

app = dash.Dash(__name__)

app.layout = html.Div([
    dcc.Input(id='transaction-amount', type='number', placeholder='Transaction Amount'),
    dcc.Input(id='transaction-zip', type='text', placeholder='Transaction Zip'),
    dcc.Input(id='transaction-city-population', type='number', placeholder='City Population'),
    # I chose these 3 as I felt they were the most relevant and easiest for someone to input as they are already numbers.
    html.Button('Predict', id='submit-button'),
    html.Div(id='prediction-output')
])

@app.callback(
    Output('prediction-output', 'children'),
    [Input('submit-button', 'n_clicks')],
    [State('transaction-amount', 'value'),
     State('transaction-zip', 'value'),
     State('transaction-city-population', 'value')]
)
def update_output(n_clicks, amount, zip_code, city_population):
    if n_clicks is None:
        return 'No prediction yet'

    prediction = predict(amount, zip_code, city_population)
    return f'The transaction is predicted to be fraudulent: {prediction}'

if __name__ == '__main__':
    app.run_server(debug=True)
```
